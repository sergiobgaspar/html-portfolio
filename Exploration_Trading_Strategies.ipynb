```python
# prompt: can you help me create a moving average of the last 50, 100, 150, 200 closing prices?

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import libraries

# Define the ticker symbol and date range
ticker = "GOOG"

# Download the historical data
data = yf.download(ticker, period="max")  # Assign the downloaded data to a variable named 'data'

# Rename the data variable to the value of ticker
globals()[ticker] = data # then assign to ticker in the global namespace.

# Print the downloaded data (optional)
print(data)

# Calculate moving averages
windows = [50, 100, 150, 200]
for window in windows:
    data[f'MA_{window}'] = data['Close'].rolling(window=window).mean()

# Print the data with moving averages
data
```
